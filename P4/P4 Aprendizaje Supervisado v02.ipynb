# 3.1 Decision Tree Classifier
con hiperparámetros por defecto. Métrica elegida: F1.

```
from sklearn.tree import DecisionTreeClassifier as DT
from sklearn.metrics import classification_report
#X_train, X_test, y_train, y_test 

tree = DT(random_state=42)

tree.fit(X_train, y_train)

y_tree_pred = tree.predict(X_test)
```

```
print(classification_report(y_test, y_tree_pred))
#al estar desbalanceado, deberiamos usar F1 com ometrica
```

```
from sklearn.metrics import plot_confusion_matrix

fig = plt.figure(figsize=(5,5))
ax = plt.subplot('111')
plot_confusion_matrix(
    tree, X_test, y_test,
    cmap=plt.cm.Blues,
    normalize='true',
    ax=ax
  )

plt.title('Confusion Matrix for DecisionTreeClassifier\nPARA TODA LA POBLACION \ndefault model')
plt.show()
```

# 3.2 Decision Tree Classifier
grid-search y 5-fold cross-validation 

```
from sklearn.model_selection import GridSearchCV


fold = 5
param_grid = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'max_features': [None, 'auto', 'sqrt', 'log2'],
    'max_depth':[None, 8, 16],
    # 'ccp_alpha': np.logspace(-1, -5, 5),
    'min_samples_split': range(2, 10),
    'min_samples_leaf':range(1,6)}

treecv = GridSearchCV(tree, param_grid, scoring='accuracy', cv=fold,n_jobs=-1)
treecv.fit(X_train, y_train);
```

```
# res_df = pd.DataFrame(treecv.cv_results_)
# cols = ['rank_test_score', 'mean_test_score', 'std_test_score', 'params']
# results = res_df[cols].sort_values(by=['rank_test_score']).head()
# results
```

```
treecv_best_model = treecv.best_estimator_
print(treecv_best_model)
```

```
y_treecv_train_pred = treecv_best_model.predict(X_train)
y_treecv_pred = treecv_best_model.predict(X_test)
```
