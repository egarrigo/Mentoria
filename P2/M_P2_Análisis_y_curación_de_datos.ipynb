
### 6) Supongamos que necesitamos compartir públicamente todo el dataset y que los features "cod_encuestador" y "dni" son sensibles y confidenciales. Elijan y apliquen dos técnicas de despersonalización para solucionar este aspecto.

```python
from hashlib import blake2b, md5
```

Tecnica de despersonalizacion 
Estrategia 1: md5 

```python
def hashit_md5(val):
    if isinstance(val, float): 
        return str(val)
    return md5(val.encode('utf-8')).hexdigest()



ds['dni_hash']=ds['dni'].apply(hashit_md5)
ds['cod_encuestador_hash']=ds['cod_encuestador'].apply(hashit_md5)
ds
#agregamos dos columnas y corroboramos los nombres generados.
```

```python
#borramos las columnas anteriores, para poder aplicar la segunda estrategía de despersonalización:
ds=ds.drop(columns=['dni_hash', 'cod_encuestador_hash'])
```


Tecnica de despersonalizacion 
Estrategia 2: blake2b 

```python
def hashit_blake2b(val):
    if isinstance(val, float): 
        return str(val)
    return blake2b(val.encode('utf-8'),digest_size=10).hexdigest()

ds['dni'] = ds['dni'].apply(hashit_blake2b)
ds['cod_encuestador'] = ds['cod_encuestador'].apply(hashit_blake2b)

#aplicamos la despersonalización directamente sobre las columnas.
ds
```

### 9) Outliers: Durante el Análisis que realizaron en nuestro Práctico anterior, descubrieron que existen valores atípicos aplicar una solución para normalizar solo esos Outliers. Recuerden que pueden consultarnos lo que haga falta!
Las variables seleccionadas fueron: sexo, IMC, eent, actfis y edad.

Sólo IMC y edad son variables continuas.

```python
#IMC
sns.set(style="darkgrid")
sns.set_context(context='talk', font_scale=1)
plt.figure(figsize=(4,6))
sns.boxplot(data=ds,y='IMC',showmeans=True, color="b")
ds.IMC.describe()
```

```python
out_imc=ds[(numpy.abs(ds.IMC - ds.IMC.mean()) >= (2.5 * ds.IMC.std()))]
print("rango de outliers:", out_imc["IMC"].min(), out_imc["IMC"].max(), "\ncantidad de outliers:",len(out_imc["IMC"]))
```

**Conclusión**: La función reconoce a los índice mayores a 37.8 como outliers, siendo que la media está 25.82. Se decide no eliminar los outliers de IMC ya que representa a la población con índice de obesidad grado II y III. Serán importantes para nuestros estudios futuros.
